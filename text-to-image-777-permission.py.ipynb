```python
import tkinter as tk
from tkinter import messagebox
from tkinter import filedialog
from PIL import Image, ImageDraw, ImageFont
import subprocess
import platform

# Function to handle button click event
def convert_to_image():
    code = code_text.get("1.0", tk.END).strip()  # Get the code from the text box
    
    if not code:
        messagebox.showerror("Error", "Please paste some code.")
        return

    # Create an image with light green background and padding
    image_width = 800
    image_height = 600
    background_color = (144, 238, 144)  # RGB light green
    padding = 10
    font_size = 16
    font_color = (0, 0, 0)  # RGB black
    font_type = ImageFont.load_default()
    image = Image.new("RGB", (image_width, image_height), background_color)
    draw = ImageDraw.Draw(image)
    
    # Set the initial position for drawing the code with padding
    x = padding
    y = padding
    
    # Draw each line of code with the same colors
    for line in code.split("\n"):
        draw.text((x, y), line, fill=font_color, font=font_type)
        y += 20
    
    # Ask user for save location and image name
    file_path = filedialog.asksaveasfilename(defaultextension=".png", filetypes=[("PNG Image", "*.png")])
    
    if file_path:
        # Save the image as PNG
        image.save(file_path)
        
        # Set file permissions based on the operating system
        if platform.system() == "Windows":
            # Grant full control access on Windows
            try:
                subprocess.run(["icacls", file_path, "/grant", "Everyone:F"], check=True)
            except subprocess.CalledProcessError:
                messagebox.showwarning("Warning", "Failed to set file permissions.")
        else:
            # Set file permissions to 777 for other systems
            try:
                subprocess.run(["chmod", "777", file_path], check=True)
            except subprocess.CalledProcessError:
                messagebox.showwarning("Warning", "Failed to set file permissions.")
        
        messagebox.showinfo("Success", f"Code converted to image and saved as {file_path}.")
    else:
        messagebox.showerror("Error", "No file selected. Code not saved.")

# Create the GUI window
window = tk.Tk()
window.title("Code to Image Converter")

# Create the text box for code input
code_text = tk.Text(window, height=20, width=80)
code_text.pack()

# Create the convert button with styling
convert_button = tk.Button(window, text="Convert to Image", command=convert_to_image, bg="lightblue", fg="black", padx=10, pady=5, font=("Arial", 12, "bold"))
convert_button.pack()

# Start the main GUI loop
window.mainloop()
```
